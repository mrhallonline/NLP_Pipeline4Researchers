```python
import json
import pandas as pd
import os

# Load the JSON file
with open('state_of_union_data/state_of_union_speeches.json', 'r') as f:  # Replace with your actual JSON filename
    data = json.load(f)

# Create a folder to save the CSV files
output_folder = "state_of_union_data/sotu_speeches"
if not os.path.exists(output_folder):
    os.makedirs(output_folder)

# Loop through each speech in the JSON data and save it as a CSV
for speech in data:
    # Extract the date and president for naming the CSV file
    date = speech.get('date', 'unknown_date')
    president = speech.get('president', 'unknown_president').replace(' ', '_')
    
    # Create a DataFrame for the speech
    df = pd.DataFrame([speech])
    
    # Generate a filename based on the date and president
    filename = f"{date}_{president}.csv"
    
    # Save the DataFrame to a CSV file
    csv_file_path = os.path.join(output_folder, filename)
    df.to_csv(csv_file_path, index=False)

    print(f"Saved: {csv_file_path}")

print("All speeches have been saved to CSV files.")
```
